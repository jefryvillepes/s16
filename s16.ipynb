```python

"""
- Si encuetras un valor - eXITO
-Si es menor -> busca la izquierda
-si es mayor -> busca la derecha
-si llegas a none -> no existe
"""
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.iz = None
        self.der = None

def insertar(raiz, valor):
    if raiz is None:
        return Nodo(valor)
    if valor < raiz.valor:
        raiz.iz = insertar(raiz.iz, valor)
    else:
        raiz.der = insertar(raiz.der, valor)
    return raiz

def buscar(raiz, valor):
    if raiz is None:
        return False
    if raiz.valor == valor:
        return True
    if valor < raiz.valor:
        return buscar(raiz.iz, valor)
    return buscar(raiz.der, valor)

raiz = None
for valor in [10, 5, 15, 3, 7]:
    raiz = insertar(raiz, valor)
    
print(f"¿Estará el 7?: {buscar(raiz, 7)}")
print(f"¿Estará el 20?: {buscar(raiz, 20)}")
print(f"¿Estará el 3?: {buscar(raiz, 3)}")

```
